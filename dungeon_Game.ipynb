```python
import random
from IPython.display import clear_output

class Player:
    def __init__(self):
        self.location = [1,1]
        self.inventory = 0
        self.basket = False
    
    def start(self):
        self.location = [random.randrange(0, 4), random.randrange(0, 4)]
        
    def move(self, move):
        if move == "w":
            if self.location[1] == 0:
                self.location[1] = 4
            else:
                self.location[1] = self.location[1] - 1
        elif move == "s":
            if self.location[1] == 4:
                self.location[1] = 0
            else:
                self.location[1] = self.location[1] + 1
        elif move == "a":
            if self.location[0] == 0:
                self.location[0] = 4
            else:
                self.location[0] = self.location[0] - 1
        elif move == "d":
            if self.location[0] == 4:
                self.location[0] = 0
            else:
                self.location[0] = self.location[0] + 1
class Monster:
    def __init__(self):
        self.location = [4,3]
        
    def start(self):
        self.location = [random.randrange(0, 4), random.randrange(0, 4)]
        
    def move(self, move):
        if move == "s":
            if self.location[1] == 0:
                self.location[1] = 4
            else:
                self.location[1] = self.location[1] - 1
        elif move == "w":
            if self.location[1] == 4:
                self.location[1] = 0
            else:
                self.location[1] = self.location[1] + 1
        elif move == "d":
            if self.location[0] == 0:
                self.location[0] = 4
            else:
                self.location[0] = self.location[0] - 1
        elif move == "a":
            if self.location[0] == 4:
                self.location[0] = 0
            else:
                self.location[0] = self.location[0] + 1
    

class Egg:
    def __init__(self):
        self.location = [0,0]
    
    def start(self):
        self.location = [random.randrange(0, 4), random.randrange(0, 4)]

class Basket:
    def __init__(self):
        self.location = [2,4]
    
    def start(self):
        self.location = [random.randrange(0, 4), random.randrange(0, 4)]
        
class Door:
    def __init__(self):
        self.location = [4,0]
    
    def start(self):
        self.location = [random.randrange(0, 4), random.randrange(0, 4)]
        

class Game:
    def __init__(self):
        self.coordinates = [
            (0,0),(1,0),(2,0),(3,0),(4,0),
            (0,1),(1,1),(2,1),(3,1),(4,1),
            (0,2),(1,2),(2,2),(3,2),(4,2),
            (0,3),(1,3),(2,3),(3,3),(4,3),
            (0,4),(1,4),(2,4),(3,4),(4,4),
        ]

    
    def draw_map(self):
        field = self.coordinates
        field_rows = [field[:5], field[5:10], field[10:15], field[15:20], field[20:25]]
        for row in field_rows:
            for cell_index in range(len(row)):
                if (monster.location[0] == row[cell_index][0]) and (monster.location[1] == row[cell_index][1]):
                    row[cell_index] = " OV-= "
                elif (player.location[0] == row[cell_index][0]) and (player.location[1] == row[cell_index][1]):
                    row[cell_index] = " Q8<= "
                elif (egg.location[0] == row[cell_index][0]) and (egg.location[1] == row[cell_index][1]):
                    row[cell_index] = "    o "
                elif (basket.location[0] == row[cell_index][0]) and (basket.location[1] == row[cell_index][1]):
                    row[cell_index] = "  C=) "
                elif (door.location[0] == row[cell_index][0]) and (door.location[1] == row[cell_index][1]):
                    row[cell_index] = "[==.=]"
                else:
                    row[cell_index] = "      "
            print(''.join(row))
    
    def events(self):
        # Flag player as having the basket and move the basket off the field
        if player.location == basket.location:
            player.basket = True
            print("You have the basket!")
            basket.location = (9, 9)
        # Add an egg to the player inventory
        if (player.location == egg.location) and (player.basket == True):
            player.inventory = player.inventory + 1
            print("You picked up an egg!")
            egg.start()
        # Move eggs from the field if the player gets 3 eggs 
        if player.inventory == 3:
            egg.location = (8,8)
        
    @classmethod
    def run(self):
        quit = False
        while not quit:
            clear_output()
            game.draw_map()
            # Win the game if player has 3 eggs and touches the door
            if (player.inventory == 3) and player.location == door.location:
                print("\nYOU WIN!")
                quit = False
                break
            # Lose the game if player and monster share a location
            if player.location == monster.location:
                print("\nYou're Dead!")
                quit = False
                break
            turn = input("What is your move?\n Use: W, A, S, D\n")
            player.move(turn)
            monster.move(turn)
            game.events()
            
                
```

```python
game = Game()
player = Player()
monster = Monster()
basket = Basket()
egg = Egg()
door = Door()
game.run()
```
